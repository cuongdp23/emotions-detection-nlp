```python
import csv

def remove_stop_words(input_file, stop_word_file):
    # Đọc dữ liệu từ file
    with open(opposite_file, 'r', newline='', encoding='utf-8') as word_csv:
        word_reader = csv.reader(word_csv)
        words = [row[0] for row in word_reader]

    # Đọc dữ liệu từ file stop_word.csv
    with open(stop_word_file, 'r', newline='', encoding='utf-8') as stop_word_csv:
        stop_word_reader = csv.reader(stop_word_csv)
        stop_words = [row[0] for row in stop_word_reader]

    # Loại bỏ các từ trong stop_words nếu có trong từ điển
    cleaned_stop_words = [word for word in stop_words if word not in words]
    with open(stop_word_file, 'w', newline='', encoding='utf-8') as stop_word_csv:
        stop_word_writer = csv.writer(stop_word_csv)
        for word in cleaned_stop_words:
            stop_word_writer.writerow([word])
```
